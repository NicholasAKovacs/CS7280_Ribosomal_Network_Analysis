```python
import json
import networkx as nx
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
with open('./file.txt', 'r') as fp:
    data = json.load(fp)
```

```python
data
```

```python
G = nx.Graph()
```

```python
for x in data:
    G.add_node(x)
    G.node[x]['x'] = data[x][0]
    G.node[x]['y'] = data[x][1]
    G.node[x]['z'] = data[x][2]
```

```python
x_coord = list(nx.get_node_attributes(G,'x').values())
y_coord = list(nx.get_node_attributes(G,'y').values())
z_coord = list(nx.get_node_attributes(G,'z').values())
```

```python
xyz_coords = []
for x,y,z in zip(x_coord, y_coord, z_coord): 
    xyz_coords.append([x, y,z])
```

```python
trace1 = go.Scatter3d(
    x=x_coord,
    y=y_coord,
    z=z_coord,
    mode='markers',
    marker=dict(
        size=12,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
